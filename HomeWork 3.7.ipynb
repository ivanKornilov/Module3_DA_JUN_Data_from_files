```python
Brands = ['Green', 'Blue', 'Orange','Yellow','Black','Deep', 'Red', 'White', 'Rosy', 'Dim', 'Pale', 'Vivid', 'Magenta']
Green_in_2022 = result[(result['Brand'] == 'Green') & (result['Year'] == 2022)]['Value (in 1000 rub)'].sum()
Green_in_2021 = result[(result['Brand'] == 'Green') & (result['Year'] == 2021)]['Value (in 1000 rub)'].sum()
dinamic_green = (1 - Green_in_2022/Green_in_2021) * 100
for _ in Brands :
    Green_in_2022 = result[(result['Brand'] == _) & (result['Year'] == 2022)]['Value (in 1000 rub)'].sum()
    Green_in_2021 = result[(result['Brand'] == _) & (result['Year'] == 2021)]['Value (in 1000 rub)'].sum()
    dinamic = (1 - Green_in_2022/Green_in_2021) * 100
    if (dinamic > dinamic_green) :
        print(f'Динамика продаж в деньгах для {_} составила {dinamic :.1f}% и лучше Green')
```
